```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("E:\RPI Courses And Stuff\Courses\Data Analytics\PROJECT\Disability_and_Health_Data_System__DHDS_.csv")
df2 = pd.read_csv("E:\RPI Courses And Stuff\Courses\Data Analytics\PROJECT\percent_of_adults_with_obesity_usafacts.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['LocationDesc'] = df1['LocationDesc'].str.lower()

df2_new  =df2.drop_duplicates(subset='Years')
df2_new['Years'] = df2_new['Years'].str.replace(r'\(.*\)', '')  # Remove everything inside parentheses
df2_new['Years'] = df2_new['Years'].str.replace(r'\s*%\s*$', '')  # Remove percentages and trim spaces
df2_new['Years'] = df2_new['Years'].str.lower()
```

```python
df2_new
```

```python
# Reshape the DataFrame using melt
df2_new_melted = pd.melt(df2_new, id_vars='Years', var_name='State', value_name='Value')
df2_new_melted.columns = ["LocationDesc","Year","Obesity_Value"]
df2_new_melted["Year"] = pd.to_numeric(df2_new_melted["Year"])
df2_new_melted['LocationDesc'] = df2_new_melted['LocationDesc'].str.strip()
df2_new_melted
```

```python
merged_data= df1.merge(df2_new_melted, on=["LocationDesc","Year"])
merged_data
```

```python
merged_data.to_csv("Data.csv")
```
